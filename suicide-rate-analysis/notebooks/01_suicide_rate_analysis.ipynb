# Global Suicide Rate Analysis

**Goal:** Analyze global suicide trends using demographic (sex, age, generation), economic (GDP), and population-level (HDI) indicators.

**Note:** This is a statistical/demographic analysis of aggregate public-health data (similar to WHO reporting), not a discussion of individual risk. All figures represent population-level rates per 100,000 people.

**Pipeline:**
1. Load & clean the data
2. Exploratory Data Analysis (EDA)
3. Trend analysis (over time)
4. Comparative analysis (by sex, age, generation, country)
5. Correlation analysis (GDP, HDI vs suicide rate)
6. Insight-driven visualization & summary

## 1. Setup & imports

```python
import sys
sys.path.append('../src')

import pandas as pd
import matplotlib.pyplot as plt

from data_loader import load_data, basic_quality_report
from analysis import (
    global_trend_by_year, top_countries_by_rate, rate_by_sex,
    rate_by_age_group, rate_by_sex_and_age, gdp_vs_suicide_rate,
    hdi_vs_suicide_rate, generation_comparison,
)
from visualize import (
    plot_global_trend, plot_top_countries, plot_rate_by_sex, plot_rate_by_age,
    plot_sex_age_heatmap, plot_gdp_scatter, plot_hdi_scatter, plot_generation_comparison,
)
```

## 2. Load & clean the data

By default this loads `data/sample_data.csv`, a small synthetic dataset included so the notebook runs out of the box.

**To do the real analysis:** download the "Suicide Rates Overview 1985 to 2016" dataset (WHO/Kaggle) and place it at `data/master.csv`, then change the path below.

```python
DATA_PATH = '../data/sample_data.csv'  # change to '../data/master.csv' for the real dataset

df = load_data(DATA_PATH)
basic_quality_report(df)
```

```python
df.head()
```

## 3. Exploratory Data Analysis

```python
print('Years covered:', df['year'].min(), '-', df['year'].max())
print('Countries:', df['country'].nunique())
print('Total recorded suicides in dataset:', df['suicides_no'].sum())
print()
df.describe()
```

```python
# Distribution of suicide rate values across all rows
plt.figure(figsize=(8, 5))
df['suicide_rate'].hist(bins=40, color='#4C72B0')
plt.title('Distribution of Suicide Rate (per 100k) across all records')
plt.xlabel('Rate per 100,000')
plt.ylabel('Frequency')
plt.show()
```

## 4. Trend Analysis: Global rate over time

```python
yearly = global_trend_by_year(df)
display(yearly)
plot_global_trend(yearly, save_path='../visuals/global_trend.png')
```

## 5. Comparative Analysis: Countries

```python
top_countries = top_countries_by_rate(df, n=10)
display(top_countries)
plot_top_countries(top_countries, save_path='../visuals/top_countries.png')
```

## 6. Comparative Analysis: Sex

```python
sex_df = rate_by_sex(df)
display(sex_df)
plot_rate_by_sex(sex_df, save_path='../visuals/rate_by_sex.png')
```

## 7. Comparative Analysis: Age Group

```python
age_df = rate_by_age_group(df)
display(age_df)
plot_rate_by_age(age_df, save_path='../visuals/rate_by_age.png')
```

## 8. Sex × Age Interaction

```python
pivot = rate_by_sex_and_age(df)
display(pivot)
plot_sex_age_heatmap(pivot, save_path='../visuals/sex_age_heatmap.png')
```

## 9. Economic Indicator: GDP per Capita vs Suicide Rate

```python
country_gdp, corr_gdp = gdp_vs_suicide_rate(df)
print(f'Correlation between GDP per capita and suicide rate: {corr_gdp:.3f}')
plot_gdp_scatter(country_gdp, corr_gdp, save_path='../visuals/gdp_scatter.png')
```

## 10. Development Indicator: HDI vs Suicide Rate

```python
country_hdi, corr_hdi = hdi_vs_suicide_rate(df)
if corr_hdi is not None:
    print(f'Correlation between HDI and suicide rate: {corr_hdi:.3f}')
    plot_hdi_scatter(country_hdi, corr_hdi, save_path='../visuals/hdi_scatter.png')
else:
    print('No HDI data available in this dataset.')
```

## 11. Generational Comparison

```python
gen_df = generation_comparison(df)
display(gen_df)
plot_generation_comparison(gen_df, save_path='../visuals/generation_comparison.png')
```
